```python
import pandas as pd
from datautils import _get_time_features,load_forecast_csv
```

```python
data, train_slice, valid_slice, test_slice, scaler, pred_lens, n_covariate_cols,data1=load_forecast_csv("ETTh1")
```

```python
import matplotlib.pyplot as plt

# 仮にdataという名前の(1, 17420, 8)形状の配列があるとします。
# 配列の形状を(17420, 8)に変更します。
data = data.reshape(17420, 8)

# 第二引数と第三引数の最後の値を選ぶ
# Pythonのインデックスは0から始まるため、2番目の列はインデックス1に、最後の列はインデックス-1になります。
column_2 = data[:, 1]
column_last = data[:, -1]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(column_2, column_last)
plt.xlabel('2nd column')
plt.ylabel('Last column')
plt.show()
```
