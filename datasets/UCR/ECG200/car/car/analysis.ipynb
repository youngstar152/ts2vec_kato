```python
import numpy as np
import matplotlib.pyplot as plt

# テキストファイルを読み込む
data = np.loadtxt('dis_base_1m_txt/03_01.txt')

data
```

```python
# プロット
plt.figure(figsize=(10, 6))
for i in range(data.shape[1]):
    plt.plot(data[:, i], label=f'Column {i+1}')

plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Multivariate Time Series Plot')
plt.legend()
plt.grid(True)
plt.show()
```
